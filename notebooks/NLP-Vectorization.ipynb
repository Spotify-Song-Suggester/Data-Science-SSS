This notebook contains code to vectorize the artist and song names. Unfortunately, we completed this too late to make it part of the app's model. But I'm including it anyway for posterity.

```
import pandas as pd

df = pd.read_pickle('df_v2.pkl')
print(df.shape)
df.head()
```

```
df = df.fillna('')

strings = []
for i in range(len(df)):
  string = df.loc[i, 'artist'] + ' ' + df.loc[i, 'track']
  strings.append(string)

strings[:5]
```

```
df2 = df.copy()

df2['strings'] = strings
df2.head()
```

```
import spacy
from spacy.tokenizer import Tokenizer

nlp = spacy.load('en_core_web_lg')

tokenizer = Tokenizer(nlp.vocab)
```

```
def get_lemmas(text):
  lemmas = []
  doc = nlp(text)

  for token in doc:
    if ((token.is_stop is False) and (token.is_punct is False) and
        (token.pos_ is not 'PRON')):
      lemmas.append(token.lemma_)

  return lemmas
```

```
df2['lemmas'] = df2['strings'].apply(get_lemmas)
df2.head()
```

```
def get_word_vectors(l):
  return [nlp(doc).vector for doc in l]

df2['vectors'] = df2['lemmas'].apply(get_word_vectors)

df2.head()
```

```
df2.to_pickle('df_vector.pkl')
```
